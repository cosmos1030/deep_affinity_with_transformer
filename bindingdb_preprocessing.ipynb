```python
# Convert IC50 from nM to M
df['IC50_M'] = df['IC50 (nM)'] / 1e9

# Truncate values to the range [10^-11 M, 10^-2 M]
df['IC50_M_truncated'] = df['IC50_M'].clip(lower=1e-11, upper=1e-2)

# Calculate pIC50
df['pIC50'] = -np.log10(df['IC50_M_truncated'])

df = df.drop(columns=['IC50 (nM)', 'IC50_M', 'IC50_M_truncated'])
df
```

```python
comp2id = {}
with open("data/vocab_compound") as f:
    for index, line in enumerate(f):
        comp2id[line.strip()] = index
print(comp2id)
```

```python
prot2id = {}
with open("data/vocab_protein") as f:
    for index, line in enumerate(f):
        prot2id[line.strip()] = index
print(prot2id)
```

```python
df = df_filtered
```

```python
import numpy as np

# Assuming df is your DataFrame

# Calculate the number of rows for the training set (70% of the dataset)
train_size = int(len(df) * 0.7)

# Shuffle the DataFrame rows if you want the split to be random
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Split the DataFrame
train_df = df_shuffled[:train_size]
test_df = df_shuffled[train_size:]
```

```python
train_df
```

```python
test_df
```

```python
# 이중리스트로 변환
train_smiles = [list(smiles) for smiles in train_df['SMILES']]
train_protein = [list(protein) for protein in train_df['Protein']]
train_pic50 = [pic50 for pic50 in train_df['pIC50']]

test_smiles = [list(smiles) for smiles in test_df['SMILES']]
test_protein = [list(protein) for protein in test_df['Protein']]
test_pic50 = [pic50 for pic50 in test_df['pIC50']]
```

```python
train_smiles2id = []
for i in train_smiles:
    temp = []
    for j in i:
        temp.append(comp2id[j])
    train_smiles2id.append(temp)

test_smiles2id = []
for i in test_smiles:
    temp = []
    for j in i:
        temp.append(comp2id[j])
    test_smiles2id.append(temp)

train_protein2id = []
for i in train_protein:
    temp = []
    for j in i:
        temp.append(prot2id[j])
    train_protein2id.append(temp)

test_protein2id = []
for i in test_protein:
    temp = []
    for j in i:
        temp.append(prot2id[j])
    test_protein2id.append(temp)
```

```python
import torch

def insert_padding(mylist, max_size):
    for i, sentence in enumerate(mylist):
        templist = [0 for _ in range(max_size-len(sentence))]
        mylist[i] = mylist[i] + templist
    return mylist

def process_independent(mylist, max_size, file_name):
    mylist = insert_padding(mylist, max_size)
    mytensor = torch.tensor(mylist)
    torch.save(mytensor, 'data2/'+file_name+".pt")
    return mytensor
```

```python
comp_MAX_size = 100
protein_MAX_size = 1000
```

```python
process_independent(train_protein2id, protein_MAX_size, "train_protein")
process_independent(test_protein2id, protein_MAX_size, "test_protein")
process_independent(train_smiles2id, comp_MAX_size, "train_smiles")
process_independent(test_smiles2id, comp_MAX_size, "test_smiles")
```

```python
def process_dependent(mylist, file_name):
    mytensor = torch.tensor(mylist)
    torch.save(mytensor, 'data2/'+file_name+".pt")
    return mytensor
```

```python
process_dependent(train_pic50, "train_pic50")
process_dependent(test_pic50, "test_pic50")
```
